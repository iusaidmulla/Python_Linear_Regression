Load the data into pandas dataframe

```python
boston = pd.DataFrame(boston_dataset['data'],columns=boston_dataset['feature_names'])
boston.head()
```

The target values is missing from the data. Create a new col. of target vals. and add it to dataframe

```python
boston['PRICE'] = boston_dataset.target
boston.head()
```

Data Preprocessing

```python
boston.isnull().sum()
```

Correlation Matrix

```python
# compute the pair wise correlation for all cols
corr_matrix = boston.corr()
corr_matrix
```

Rounding off to two decimal places

```python
corr_matrix = boston.corr().round(2)
corr_matrix
```

Use the heatmap function from seaborn to plt the correlation matrix

```python
# annot = true to print the values inside the square
plt.figure(figsize=(10,10))
sns.heatmap(data=corr_matrix , annot=True)
```

Observation

RAD and TAX are strongly correlated, so we don't include this in our features together to avoid multi-colinearity

```python
boston
```

```python
y = boston['PRICE']
```

```python
X = boston.drop(['PRICE','TAX'],axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.33, random_state=5)
print(X_train.shape) # 70% data of all the feature except price dependent variable
print(X_test.shape)  # 30% data of all the feature except price dependent variable
print(Y_train.shape) # 70% data of price feature dependent variable
print(Y_test.shape)  # 30% data of price feature dependent variable
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

```python
# y = mx+c
lm.coef_ # m
```

```python
lm.intercept_ # c
```

```python
coef_df = pd.DataFrame(lm.coef_).T
coef_df.columns = X_train.columns

coef_df['C'] = lm.intercept_
coef_df
```

```python
plt.figure()
coef_df.plot(kind='bar')
plt.show()
```

```python
Y_pred = lm.predict(X_test)
Y_pred
```

Lets compare the predicted with actual

```python
Y_test[:5]
```

```python
pd.DataFrame(Y_pred[0:5])
```

Assess the model

```python
from sklearn import metrics
```

```python
metrics.r2_score(Y_test,Y_pred)
```

```python
# Mean Absol. Error
metrics.mean_absolute_error(Y_test,Y_pred)
```

```python
# Mean Squared Error
metrics.mean_squared_error(Y_test,Y_pred)
```

```python
# Root Mean Squared Error
np.sqrt(metrics.mean_squared_error(Y_test,Y_pred))
```
